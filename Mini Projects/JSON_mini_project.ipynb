It seems there can be two approaches (as project can have more than one theme).
A) First one is, that we group project by name (this means that projects can be counted as times as many themes they have --> 1 project = many themes).
B) Second is, that project with multiple theme, we choose the first one as a major (this means that we choose one theme for project
and therefore 1 project = 1 theme).
I think both approach are correct and which one to choose purely depends what the author of the exercise wanted - and in business what the contractor wants). So I'll do both here.

```python
# A) Here grouping by 'name', then counting how many times 'name' project has occured and then sort it.
by_name = df_project.groupby('name')
count_by_name = by_name['name'].count().sort_values(ascending=False).head(10)
count_by_name
```

```python
#B) Here grouping by 'id', counting how many times name' project has occured.
by_id = df_project.groupby('id').first()
count_by_name_B = by_id.name.value_counts().head(10)
count_by_name_B
```
